## 1. Using any finance API (we use Yahoo Finance and Tiingo during sessions) download historical stock prices for Google (symbol ‘GOOG’), for the period 1 Jan 2020 to 1 Jan 2021.

```python
# Tingo
# # ! pip install tiingo
from tiingo import TiingoClient

# data manipulation
import csv
import pandas as pd

# plotting
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
token = 'e61e7ea7f32831962ca13f1cf0d631ecd7000d30'
client = TiingoClient({'api_key': token})
```

```python
google_data = client.get_dataframe('GOOGL',
                                    frequency='daily',
                                    #metric_name='close',
                                    startDate='2020-01-01',
                                    endDate='2021-01-01')


# google_data.head()
```

## 2. Save these prices to csv file with Pandas

```python
google_data.to_csv('google_data_raw.csv')
```

## 3. Read this csv file to display data saved.

```python
google_data = pd.read_csv('google_data_raw.csv') # this step is redundant as google data already exists as a pandas dataframe
```

## 4. Plot a simple diagram to display closing prices for GOOG stock.

```python
google_data['date'] = mdates.date2num(google_data['date']) # converts datetime object to matplotlib date

month_locator = mdates.MonthLocator(interval = 1)
year_month_formatter = mdates.DateFormatter("%B") # only month names

fig, ax = plt.subplots(figsize=(20, 10)) # setting plot size
ax.xaxis.set_major_locator(month_locator) # Locator for major axis only.
ax.xaxis.set_major_formatter(year_month_formatter)

plt.plot(google_data['date'], google_data['close'], color = '#780a21') # plotting 2 variables
plt.ylim(0,2000) # y axis starts at 0

plt.xlabel('Date', fontsize = 20) # labelling x axis
plt.ylabel('$', fontsize = 20) # labelling y axis
plt.title('Closing prices for Google Stock \nJanuary 2020 - January 2021 ', fontsize = 20) #plot title

plt.show() # visualize
```
