```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris_data = pd.read_csv('iris.csv')
```

### 1. GRAPH 1: Create a graph to find relationship between the sepal length and width. 
      

```python
fig = iris_data[iris_data.species=='setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='orange', label='Setosa')
iris_data[iris_data.species=='versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='blue', label='versicolor',ax=fig)
iris_data[iris_data.species=='virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

- Describe and interpret graph results. 

**scatter plot**<br>
This scatter plot displays the relationship between two variables by plotting each data point. The position ofeach data point is determined by the value of x and y, here sepal length and sepal width. Here additional information is encodeda s color of the data point which gives information about the iris species.

**Descriptive stats**<br>
The length of sepals in the iris dataset ranged between approximately 4.3 - 8 cm and the width of sepals ranged between
2.2 - 4 cm. On average, the setosa species has longer and narrower sepals than to virginica and versicolor species. While the versicolor and virginica species have similar length of sepals,iris virginica appears to have marginally wider sepals. In terms of length, the virginica species has longer sepals than both the setosa and versicolor. 

**sepal length vs. width relationship**<br>
There is a positive correlation between the sepal width and sepal length of the flowers for all three species. In other words, the longer the sepal is, the wider it will be. While the positive linear relationship between width and length of teh sepals is true for all three iris species, the relationship is the strongest (i.e. the linear regression slope is the highest) for the setosa species. This means that as the sepal length increases the sepal width will increase faster in iris setosa compared to iris virginica and iris versicolor.

- Describe and interpret graph results.

**Hex jointplot**<br>
A jointplot  hexplot is a plot of two variables (here sepal length on the x axis, and sepal width on the y axis) with bivariate and univariate graphs. The univariate graphs are histograms plotted at the top for the x axis variable (sepal length) and far right for the y axis variable (sepal width). These histograms show the distribution of the individual measurements. The bivariate hexplot graph is similar to a scatterplot described above, but instaed of plotting each data point we plot a hexbin which color is determined by the number of datapoints inside of it, ie.teh density. The more data points fall within a given hexbin, teh darker the color.

**Univariate: sepal length and sepal width**<br>
Looking at the histograms, the distribution of sepal  width is close to a normal distributions, with some bias towards shorter sepal widths. A sepal width is more likely to be shorter than a mean. The skew of the sepal length variable is more noticable. The sepal width is also skewed towards shorter sepal lengths. That is, there are fewer sepals with longer than a mean sepal length than thereare petals with shorter than a mean length.
 
**Bivariate: sepal length vs. width relationship**<br>
The hexplot displays the relationship between length and width of the sepals in irises and provides similar information to the scatterplot described and interpreted above.<br>
 
The hexplot shows the counts of observations that fall within the hexagonal bins. We can see an indication of a positive linearrelationship between sepal length and sepal width, however this ismore difficult to see than in the scatterplot above. <br><br>
Closer inspection of the hexplot and the darkest hexbins (those with highest density of datapoints) suggests there are two clusters of plants. While it is impossible to infer any more information about the clusters from the hexplot alone, if we consider the scatterplot above we can see that these clusters correspond to the setosa species on one hand and virginica and versicolor on the other.<br>
