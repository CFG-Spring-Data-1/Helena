```python
import datetime
import pandas_datareader as pdr
import pandas as pd

import matplotlib.pyplot as plt 
import seaborn as sns


api_key='e61e7ea7f32831962ca13f1cf0d631ecd7000d30'
```

```python
hw_tickers = ['PFE', 'MRNA', 'AZN', 'JNJ']

# PFE - pfizer
# MRNA - Moderna
# AZN - astrazeneca
# JNJ - Johnson & Johnson


start="2020-1-1"
end="2021-7-1"

hw_df = pdr.tiingo.TiingoDailyReader(hw_tickers, start=start, end=end, api_key=api_key)


hw_data = hw_df.read()
# hw_data.head(10)
```

## Original Histograms

```python

# Isolate the `Adj Close` values and transform the DataFrame
hw_daily_close_px = hw_data[['adjClose']].reset_index().pivot('date', 'symbol', 'adjClose')

# Calculate the daily percentage change for `daily_close_px`
hw_daily_pct_change = hw_daily_close_px.pct_change()

# change column headings
hw_daily_pct_change.columns = ['Astra Zeneca', 'Johnson & Johnson', 'Moderna', 'Pfizer']

# Plot the distributions
hw_daily_pct_change.hist(bins=50, 
                         sharex=True,
                         sharey=True, 
                         figsize=(12,8), 
                         color = 'Hotpink', 
                         alpha=0.75)




# Show the resulting plot
plt.show()
```

## Histograms with a kernel density estimate 

```python
for company in hw_daily_pct_change.columns[0:4]:

    plt.figure(figsize = (6,8))
    sns.displot(hw_daily_pct_change[company], bins = 50, alpha =0.5 )
    sns.kdeplot(data = hw_daily_pct_change[company], color = 'red' )
    plt.ylim(0,50)

```

## Time series plot with a 30-day rolling average smoothing

```python
# 30-day rolling average to smooth the lines
for company in hw_daily_pct_change.columns[0:4]:
    hw_daily_pct_change['roll ' + company] = hw_daily_pct_change[company].rolling(30).mean().shift(-3)
```

```python
# Time series plots with 30-day rolling average in red to see what is happening over time
for company in hw_daily_pct_change.columns[0:4]:

    hw_daily_pct_change[company].plot()
    hw_daily_pct_change['roll '+ company].plot(color = 'red')
    plt.xlabel('date')
    plt.ylabel('daily% change')
    plt.title(company +': \nDaily percentage change of stock prices')
    plt.show()

```

## Data Interpretation


I looked at the daily % change of closing stock prices for Moderna, Johnson&Johnson, Pfizer and AstraZeneca pharmaceutical companies over the period between 1/1/2020 and 1/7/2021. Positive values of % change 
indicate increase in stock price compared to previous day, while negative values indicate a decrease of stock prices.

Looking at the histograms we can see that the values are close to a normal distribution and thus symmetrical. This is apparent from the first set of histograms (pink) and is even clearer when we examine the histograms with overlay of kernel density plots (blue).  As the distribution does not appear to be skewed to either positive or negative values, this means that the daily change in stock prices was sometimes up, sometimes down, but overall there was  a balance between daily % increases and decreases. 

The most dtriking difference between the four pharma companies is that the distribution of daily % change for Johnson&Johnson, Astra Zeneca and Pfizer is relatively narrow compared to the distribution of daily % change for Moderna. The wider score distribution in daily % change for Moderna indicates that Moderna's stock proces were realtively less stable, i.e. fluctuated more compared to the other three Pharma companies.More specifically, while Moderna daily % change of stock prices fluctuated between +/- 30%, the fluctuation for the other comapnies largely stays within +/- 10%.

While the day-to-day stock prices were fluctauting widely, as is apparent both from the histograms and more so from the time-series plots, there seemed to be as many decreases as increases in the longterm, which is clear from the 30-day rolling average line plotted on the time series plot.

Wecan also notice that the period of biggest changes for all four companies, and a relatively slight skew towards growth occured around March-May 2020, the emergence  of covid-19 pandemic.

That said, there are limitations what daily % change in stock prices can tell us. For example, we cannot tell the absolute changes in stock prices. This analysis would have to be followed up by more indepth analysis to gain more meaningful insight.
