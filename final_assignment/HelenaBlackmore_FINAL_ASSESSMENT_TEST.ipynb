### Q1: What is Machine Leanring?
- machine learning is a subset of AI (artificial intelligence)
- the idea of machine learning is to feed data (usually large datasets) into a computer and let it detect patterns
    - these patterns can be used for classification (and other predictions) or clustering 
- ML canbe supervised or unsupervised, depending on the input data

### Q2: What are the key principles of 'Supervised' Machine learning?
- supervised ML requires labelled data
- this means that we supply data and their labels (e.g. cat/dof for an image, or True/False when predicting a salary)
- features in teh datasetare then used to detect patterns which predict whether the outcome
- we usually need to split our data set into training and test dataset (training dataset is used for training, i.e. detecting patterns and weighting the influence of different features), while testing dataset(s) is then used for model validation

### Q3: With regards to Machine Learning Model Fitting, what do we mean my Underfitting and Overfitting?
- **overfitting**
    - occurs when our model is unnecessarily complex and as a result mistakes noise for data
    - overfitted model will perform very well on teh training dataset, but poorly on testing dataset
    - overfitted models don't generalise well
- **underfitting**
    - when model does not detect patterns in data well enough 
    - this could happen if we don't use all informative features (independent variables) in the dataset to build teh model
    - performs relatively poorly in both training and testing dataset


### Q4: Explain how do we ususally split data in preparation for the Machine Learning process?
- (after cleaning!) data is divided into 2 or 3 subsets
    - typically we split data into training and test data set
    - sometimes we can have also validation dataset to further validate the ML
    - the proportion between test and train data varies, but 80% for training and 20% for testing is not unusual

### Q5: What is SciKit?
- SciKit Learn (sklearn) is a Python library that has number of modules useful for modelling and machine learning
- we can use modules from sklearn linear regression, non-negative matrix factorisation or supervised classification (logistic regression)
